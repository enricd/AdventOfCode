```python
with open("data_inputs/day05_input.txt") as f:
	data = f.read().split("\n")

fresh_ranges = []
ingredients = []

for line in data:
	if line:
		if "-" in line:
			fresh_ranges.append(tuple(map(int, line.split("-"))))
		elif len(line) > 2:
			ingredients.append(int(line))

fresh_ranges = [(range_i[0], range_i[1]+1) for range_i in fresh_ranges]

len(data), fresh_ranges[:3], len(fresh_ranges), ingredients[:3], len(ingredients)
```

```python
# ---- Part 1 ----

count_fresh = 0

for ingredient in ingredients:
    for range_i in fresh_ranges:
        if ingredient in range(range_i[0], range_i[1]):
            count_fresh += 1
            break
             
        
print("Result 1:", count_fresh)
```

```python
# ---- Part 2 ----

is_change = True
fresh_ranges2 = fresh_ranges.copy()

while is_change:
	is_change = False
	for i, range_i in enumerate(fresh_ranges2):
		for j, range_j in enumerate(fresh_ranges2):
			if range_i != range_j:
				if range_i[0] > range_j[1] or range_i[1] < range_j[0]:    	# not overlapping
					pass
				elif range_i[0] >= range_j[0] and range_i[1] <= range_j[1]: # full overlap j over i
					fresh_ranges2.remove(range_i)
					is_change=True
					break
				elif range_j[0] <= range_i[0] < range_j[1]:					# partial overlap j left and i right
					new_range = (range_j[0], range_i[1])
					fresh_ranges2.remove(range_i)
					fresh_ranges2.remove(range_j)
					fresh_ranges2.append(new_range)
					is_change=True
					break

	
fresh_ranges2 = list(set(fresh_ranges2))

print("Result 2:", sum([r[1]-r[0] for r in fresh_ranges2]))																	  
        
```
