```python
with open("./data_inputs/day03_input.txt") as f:
    data = f.readlines()
    
batt_banks = [line.strip("\n") for line in data]
batt_banks[:5], len(batt_banks)
```

```python
# ---- Part 1 ----

max_jolts = []

for bank in batt_banks:
    n1 = max(bank[:-1])
    n1_pos = bank.find(n1)
    n2 = max(bank[n1_pos+1:])
    max_jolts.append(int(n1 + n2))
    

print("Result 1: ", sum(max_jolts))  
```

```python
# ---- Part 2 ----

max_jolts = []

for bank in batt_banks:
	last_pos = 0
	curr_jolt = []
	for i in range(12):
		n = max(bank[last_pos : (-(11-i) or None)])
		last_pos = bank[last_pos : (-(11-i) or None)].find(n) + last_pos + 1
		curr_jolt.append(n)

	max_jolts.append(int("".join(curr_jolt)))


print("Result 2: ", sum(max_jolts))  
```
