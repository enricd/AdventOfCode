```python
with open("data_inputs/day04_input.txt") as f:
	data = f.readlines()

rolls_map = [[int(x == "@") for x in line.strip("\n")] for line in data]
len(rolls_map), len(rolls_map[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt

rolls_arr = np.array(rolls_map)

plt.figure(figsize=(10, 10))
plt.imshow(rolls_arr)
plt.show()
```

```python
# ---- Part 1 ----

def get_adj_matrix_rolls(pos_i, pos_j, max_i=len(rolls_map), max_j=len(rolls_map[0])):
	rolls_in_matrix = 0
	for ii in range(max(pos_i-1, 0), min(pos_i+2, max_i)):
		for jj in range(max(pos_j-1, 0), min(pos_j+2, max_j)):
			if pos_i == ii and pos_j == jj:
				continue
			rolls_in_matrix += rolls_map[ii][jj]
    
	return rolls_in_matrix
            

accessible_rolls = 0

for i in range(len(rolls_map)):
	for j in range(len(rolls_map[0])):
		if rolls_map[i][j]:
			if get_adj_matrix_rolls(i, j) < 4:
				accessible_rolls += 1

print("Result 1:", accessible_rolls)  
```

```python
# ---- Part 2 ----
            
accessible_rolls = 0
is_rolls_in_prev_round = True

while is_rolls_in_prev_round:
	is_rolls_in_prev_round = False
	for i in range(len(rolls_map)):
		for j in range(len(rolls_map[0])):
			if rolls_map[i][j]:
				if get_adj_matrix_rolls(i, j) < 4:
					rolls_map[i][j] = 0
					accessible_rolls += 1
					is_rolls_in_prev_round = True


print("Result 1:", accessible_rolls)  
```
