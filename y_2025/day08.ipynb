```python
with open("data_inputs/day08_input.txt", "r") as f:
    data = f.read().split("\n")

boxes = [tuple(map(int, line.split(","))) for line in data]
boxes[:5], len(boxes)
```

```python
# ---- Part 1 ----

import math

def euclidean_dist(p1, p2): 
    x1, y1, z1 = p1
    x2, y2, z2 = p2
    return math.sqrt((x1-x2)**2 + (y1-y2)**2 + (z1-z2)**2)


boxes_distances = {}

for box1 in boxes:
    boxes2 = boxes.copy()
    boxes2.remove(box1)
    for box2 in boxes2:
        if (box2, box1) not in boxes_distances:
            boxes_distances[(box1, box2)] = euclidean_dist(box1, box2)

top_1000_pairs = sorted(boxes_distances, key = lambda box_pairs: boxes_distances[box_pairs])[:1000]
```

```python
boxes_conns = {box: set() for box in boxes}
paired_boxes = set()

for pair in top_1000_pairs:
    boxes_conns[pair[0]].add(pair[1])
    boxes_conns[pair[1]].add(pair[0])
    paired_boxes.add(pair[0])
    paired_boxes.add(pair[1])

visited = set()
circuits = []

for box in paired_boxes:
    if box in visited:
        continue
    stack = [box]
    circuit_i = set()
    while stack:
        box2 = stack.pop()
        if box2 in visited:
            continue
        visited.add(box2)
        circuit_i.add(box2)
        for neighbour_box in boxes_conns[box2]:
            if neighbour_box not in visited:
                stack.append(neighbour_box)
    
    circuits.append(circuit_i)


print("Result 1:", math.prod(sorted([len(circuit) for circuit in circuits], reverse=True)[:3]))
                                                                                    
```

```python
# ---- Part 2 ----

sorted_dist_pairs = sorted(boxes_distances, key = lambda box_pairs: boxes_distances[box_pairs])

circuits = []

for pair in sorted_dist_pairs:
    box1, box2 = pair

    pair1_circuit, pair2_circuit = None, None
    for circuit in circuits:
        if box1 in circuit:
            pair1_circuit = circuit
            break
    for circuit in circuits:
        if box2 in circuit:
            pair2_circuit = circuit
            break
    
    if not pair1_circuit and not pair2_circuit:
        circuits.append([box1, box2])
    elif not pair1_circuit:
        pair2_circuit.append(box1)
    elif not pair2_circuit:
        pair1_circuit.append(box2)
    elif pair1_circuit and pair2_circuit:
        if pair1_circuit != pair2_circuit:
            new_circuit = pair1_circuit + pair2_circuit
            circuits.remove(pair1_circuit)
            circuits.remove(pair2_circuit)
            circuits.append(new_circuit)

    if len(circuits) == 1 and len(circuits[0]) == len(boxes):
        break


print("Result 2:", box1[0]*box2[0])  
```
