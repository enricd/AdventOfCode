```python
with open("./data_inputs/day01_input.txt", "r") as f:
    data = f.readlines()

rotations = [line.strip("\n") for line in data]

rotations[:5], len(rotations)
```

```python
# ---- Part 1 ----

pos = 50
count0 = 0

for i, rot in enumerate(rotations):
    dir, n = rot[0], int(rot[1:])

    if dir == "R":
        pos += n

    elif dir == "L":
        pos -= n
        if pos < 0:
            pos = 100 + pos

    pos = pos % 100

    if pos == 0:
        count0 += 1
    

print("Result 1:", count0)
```

```python
# ---- Part 2 ----

pos = 50
count0 = 0

for i, rot in enumerate(rotations):
    dir, n = rot[0], int(rot[1:])

    if dir == "R":
        pos += n
        count0 += pos // 100
        pos = pos % 100

    elif dir == "L":
        if pos == 0: count0 -= 1
        pos -= n

        if pos < 0:
            count0 += abs(pos // 100)
            pos = pos - (pos//100 * 100)

        if pos == 0:
            count0 += 1


print("Result 2:", count0)  
```
