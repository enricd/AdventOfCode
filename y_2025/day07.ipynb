```python
with open("data_inputs/day07_input.txt", "r") as f:
    data = f.read().split("\n")

data[:5], len(data), len(data[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt

diagram_map = {".": 0, "^": -1, "S": 1}

diagram = [list(map(lambda x: diagram_map[x], list(line))) for line in data]
```

```python
diagram_array = np.array(diagram)
plt.figure(figsize=(6, 6))
plt.imshow(diagram)
plt.axis('off')
plt.show()
```

```python
# ---- Part 1 ----

split_count = 0

for i, line in enumerate(diagram[:-1]):
    for j, pos in enumerate(line):
        if pos == 1:
            if diagram[i+1][j] == -1:
                diagram[i+1][j-1], diagram[i+1][j+1] = 1, 1
                split_count += 1
            else:
                diagram[i+1][j] = 1
                

print("Result 1:", split_count)
```
