```python
with open("data_inputs/day09_input.txt", "r") as f:
    data = f.read().split("\n")

tiles = [tuple(map(int, line.split(","))) for line in data]
tiles[:5], len(tiles)
```

```python
# ---- Part 1 ----

def area(t1, t2):
    return (abs(t1[0]-t2[0]) + 1) * (abs(t1[1]-t2[1]) + 1)

max_area = 0

for tile1 in tiles:
    for tile2 in tiles:
        if (curr_area := area(tile1, tile2)) > max_area:
            max_area = curr_area


print("Result 1:", max_area)  # low 4749534413
```
