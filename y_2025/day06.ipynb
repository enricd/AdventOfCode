```python
with open("data_inputs/day06_input.txt") as f:
	data = f.read()
	for i in range(8, 1, -1):
		data = data.replace(" "*i, " ")
		print(i, data.count(" "*i), end="   ")
	
ops = [line.split(" ") for line in data.split("\n")]

for i in range(5):
    while "" in ops[i]:
        ops[i].remove("")
    if i < 4:
        ops[i] = list(map(int, ops[i]))

len(ops), len(ops[0]), len(ops[1]), len(ops[2]), len(ops[3]), len(ops[4]), set(ops[4])
```

```python
# ---- Part 1 ----

total_1 = 0

for i in range(len(ops[0])):
    op = ops[4][i]
    if op == "+":
        result_i = sum([ops[j][i] for j in range(4)])
    else:
        result_i = ops[0][i] * ops[1][i] * ops[2][i] * ops[3][i]

    total_1 += result_i

print("Result 1:", total_1)
```

```python
# ---- Part 2 ----

with open("data_inputs/day06_input.txt") as f:
	data = f.read().split("\n")
	
total_2 = 0
nums_i = []

for i in range(len(data[0])-1, -1, -1):
    if nums_i is not None:
        nums_i.append(int("".join([data[j][i] for j in range(4)])))
        if (op := data[4][i]) != " ":
            if op == "+":
                result_i = sum(nums_i)
            else:
                result_i = 1
                for n in nums_i:
                    result_i *= n
                
            total_2 += result_i
            nums_i = None
    
    else: 
        nums_i = []


print("Result 2:", total_2)    
```
