```python
with open("./data_inputs/day02_input.txt") as f:
    data = f.read()
    
ids = [[*map(int, line.split("-"))] for line in data.split(",")]
ids[:5], len(ids)
```

```python
# ---- Part 1 ----

result1 = 0

for n0, n1 in ids:
	for n in range(n0, n1+1):
		n_str = str(n)
		if len(n_str) % 2 == 0:
			if n_str[:len(n_str)//2] == n_str[len(n_str)//2:]:
				result1 += n


print("Result 1:", result1)
```

```python
# ---- Part 2 ----

result2 = 0

for n0, n1 in ids:
	for n in range(n0, n1+1):
		n_str = str(n)
		n_len = len(n_str)
		for i in range(1, n_len//2+1):
			if n_len % i == 0:
				if n_str[:i] * (n_len//i) == n_str:
					result2 += n
					break


print("Result 2:", result2) 
```
