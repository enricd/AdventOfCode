```python
with open("data_inputs/day06_input.txt") as f:
    data = f.read()

# ---- Part 1 ----

total_times = [int(n) for n in data.split("\n")[0].split(":")[1].split(" ") if n.isdigit()]
records = [int(n) for n in data.split("\n")[1].split(":")[1].split(" ") if n.isdigit()]
```

```python
def calc_possible_distances(total_time):
    distances = []
    for press_time in range(total_time):
        v_i = press_time
        remaining_time = total_time - press_time
        distances.append(v_i * remaining_time)

    return distances


nums_record_distances = []
for i, total_time in enumerate(total_times):
    distances_i = calc_possible_distances(total_time)
    nums_record_distances.append(len(list(filter(lambda x: x > records[i], distances_i))))

result1 = 1
for num in nums_record_distances:
    result1 *= num
    
print("Result 1:", result1)
```
